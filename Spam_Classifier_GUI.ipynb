```python
import string

def preprocess_text(message):

    """
    Takes in a string of text, then performs the following:
    1. Remove all punctuation
    2. Remove all stopwords
    3. Returns a list of the cleaned text

    """

    without_punc = [char for char in message if char not in string.punctuation] #CHECK_CHARACTERS_TO_SEE_IF_THEY_ARE_IN_PINCTUATION

    without_punc = ''.join(without_punc) #JOIN_THE_CHARACTERS_AGAIN_TO_FORM_THE_STRING

    return [word for word in without_punc.split() if word.lower() not in stopwords.words('english')] #NOW_JUST_REMOVE_ANY_STOPWORDS
```

# **COUNT-VECTORIZER BASICS**

```python
from sklearn.feature_extraction.text import CountVectorizer
x = df['message']
y = df['label']
cv = CountVectorizer()
x= cv.fit_transform(x)
```

# **TEST-TRAIN MODEL**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=0)
```

```python
from sklearn.naive_bayes import MultinomialNB
classifier = MultinomialNB().fit(x_train, y_train)
```

# **ACCURACY-SCORE-AND-CONFUSION-MATRIX**

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
pred = classifier.predict(x_test)
print(classification_report(y_test, pred))
print()
print('Confusion Matrix:\n',confusion_matrix(y_test, pred))
print()
print('Accuracy : ',accuracy_score(y_test, pred))
```

```python
print(classifier.predict(x_test)) #PRINT_THE_PREDICTIONS

print(y_test.values) #PRINT_THE_ACTUAL_VALUES
```

# **FINAL-CHECK**

```python
def sms(text):

    lab = ['not spam','spam'] #CREATING_LIST_OF_LABELS

    x = cv.transform(text).toarray() #PERFORM_TOKENIZATION

    p = classifier.predict(x) #PREDICTING_TEXT

    s = [str(i) for i in p] #CONVERT_THE_WORDS_IN_STRING_WITH_THE_HELP_OF_LIST
    a = int("".join(s))

    res = str("This message is looking: "+ lab[a]) #SHOW_OUT_THE_FINAL_RESULT
    print(res)
```
